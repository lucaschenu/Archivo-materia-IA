```
columns_to_drop = []
```

```
for col in df_clean.columns:
  if df_clean[col].isnull().sum()>len(df_clean)*0.0:
    columns_to_drop.append(col)
  elif df_clean[col].dtype =="object":
    columns_to_drop.append(col)
```

```
# Revert the change that converted 'TIME' to numeric
# Iterate over all columns except 'TIME' and convert them to numeric
for col in df_clean.columns:
  if col != 'TIME':
    df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")
```

```
df_clean=df_clean.dropna(subset=["timestamp"])
```

```
target="timestamp"
```

```
features=[col for col in df_clean.columns if col not in [target]]
```

```
features
```

```
x= df_clean [features]
y=df_clean [target]
```

```
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```
from sklearn.linear_model import LinearRegression
```

```
sencillo = LinearRegression()
```

```
sencillo.fit(x_train, y_train)
```

```
print("Coefficients:", sencillo.coef_)
```

```
print("Intercept:", sencillo.intercept_)
np.float64(sencillo.intercept_)
```

```
from sklearn.metrics import r2_score, mean_absolute_error , root_mean_squared_error
```

```
y_pred_test = sencillo.predict(x_test)
```

```
r2 = r2_score(y_test, y_pred_test)
print("R-squared:", r2)
```

```
mae = mean_absolute_error(y_test, y_pred_test)
print("Mean Absolute Error:", mae)
```

```
rmse = root_mean_squared_error(y_test, y_pred_test)
print("Root Mean Squared Error:", rmse)
```
